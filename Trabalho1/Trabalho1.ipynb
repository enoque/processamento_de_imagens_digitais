```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("seagull.png", cv2.IMREAD_GRAYSCALE)
h1 = np.array([[0,0,-1,0,0],[0,-1,-2,-1,0], [-1,-2,16,-2,-1], [0,-1,-2,-1,0], [0,0,-1,0,0]])
altura, largura = h1.shape
pad = (altura)//2
img = cv2.copyMakeBorder(img, pad, pad, pad, pad, cv2.BORDER_REPLICATE)
opencvOutput = cv2.filter2D(img, -1, h1)
plt.imshow(opencvOutput, "gray")
```

# Filtragem da imagem com a máscara h1

```python
h2 = np.array([[1,4,6,4,1],[4,16,24,16,4], [6,24,36,24,6], [4,16,24,16,4], [1,4,6,4,1]])
h2 = (1/256)*h2
altura, largura = h1.shape
pad = (altura)//2
img = cv2.copyMakeBorder(img, pad, pad, pad, pad, cv2.BORDER_REPLICATE)
opencvOutput = cv2.filter2D(img, -1, h2)
plt.imshow(opencvOutput, "gray")
```

# Filtragem da imagem com a máscara h2

```python
img = cv2.imread("seagull.png", cv2.IMREAD_GRAYSCALE)
img_aux = cv2.imread("seagull.png", cv2.IMREAD_GRAYSCALE)
h3 = np.array([[-1,0,1],[-2,0,2], [-1,0,1]])
h4 = np.array([[-1,-2,-1],[0,0,0], [-1,-2,-1]])
altura, largura = h3.shape
pad = (altura)//2
img = cv2.copyMakeBorder(img, pad, pad, pad, pad, cv2.BORDER_REPLICATE)
img_aux = cv2.copyMakeBorder(img_aux, pad, pad, pad, pad, cv2.BORDER_REPLICATE)
opencvOutput = cv2.filter2D(img, -1, h3).astype('uint32')
opencvOutput2 = cv2.filter2D(img_aux, -1, h4).astype('uint32')
opencvOutput = (opencvOutput**2) + (opencvOutput2**2)
opencvOutput = np.sqrt(opencvOutput)
plt.imshow(opencvOutput.astype('uint32'), "gray")
```
